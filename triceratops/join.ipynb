```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("processed_final.csv")
```

```python
df_triceratops = pd.read_csv("triceratops_classified.csv", 
                             usecols=["TCE_ID", "triceratops_probability", "triceratops_disposition"])
```

```python
df_triceratops = df_triceratops[df_triceratops["triceratops_disposition"].notnull()]
```

```python
df_triceratops.head(10)
```

```python
df["TCE_ID"] = df.apply(lambda x: str(x[0]).zfill(9) + "-" + str(x[1]).zfill(2), axis=1)
```

```python
df.head(10)
```

```python
df_joined = df.set_index("TCE_ID").join(df_triceratops.set_index("TCE_ID"), on="TCE_ID", how="left", lsuffix='_w2p', rsuffix='_tri')
```

```python
df_joined.head(10)
```

```python
df_joined[df_joined["triceratops_disposition"].notnull()]
```

```python
df_joined.to_csv("processed_final_with_triceratops.csv", index=False)
```
