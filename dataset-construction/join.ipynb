```python
import pandas as pd
import numpy as np
from astroquery.mast import Observations
```

```python
def find_tic(kic):
    # supply KIC
    # restrict the mission to TESS
    # had to search by radius. Only produced results other than FFI around 0.2 degrees 
    # How valid is this join?
    objectname = "KIC " + str(kic)
    obs_table = Observations.query_criteria(objectname=objectname, radius=".2 deg", obs_collection="TESS")
    obs_table = obs_table[obs_table["target_name"]!="TESS FFI"] # remove "TESS FFI results"
    if len(obs_table) > 0:
        return obs_table[0]["target_name"]
    else:
        return "N/A"
```
